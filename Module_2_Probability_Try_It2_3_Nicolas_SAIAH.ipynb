In Python, there are numberous functions you can use to find probability distribution.

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
from scipy import stats
import math
```

In this exercise, you will use stats.norm() to generates continuous random data. The parameter scale refers to standard deviation and variable loc refers to mean. plt.show() is used to visualize the data. Attenpt lowering the variable "n" to see how this effects the results.

```python
# Expected Total Demand
mean = 800

#Calculate Standard Deviation
std = 250

#We have a guarenteed supply of 1000 units
guaranteedSupply = 1000

#Number of samples. See what happens when you lower this variable
n = 1000

#Set the line space going from 0 to 1600 with 1000 samples
x = np.linspace(0, 1600, n)

#Find the probability density function
pdf_result = stats.norm.pdf(x,loc=mean,scale=std)

#Label and plot the PDF against the pre-defined line space
plt.plot(x,pdf_result)
plt.xlabel('Demand')
plt.ylabel('Density')
plt.title("PDF of a Normal Distribution with mean=800 and SD=250")
plt.show()
```

Next let us calculate the Cumulative Desity Function to see what the fulfillment rate will be if we have 1000 units in inventory

```python
cdf = stats.norm.cdf(guaranteedSupply,mean,std)
cdf
```

Next lets find the Percentile point function (CDF inverse) to find how many units will be needed to have a 95% fillment rate.

```python
ppf = stats.norm.ppf(0.95,mean,std)
ppf
```

Finally let's plot the Percentile point function against our PDF to visualize the results.

```python
#Plot the PPF as a line
plt.axvline(ppf)
plt.plot(x,pdf_result)

plt.xlabel('Demand')
plt.ylabel('Density')
plt.title("PDF of a Normal Distribution with mean=800 and SD=250")
plt.show()
```

Additionally, you can greate a bar graph for a different visualization

```python
results=stats.norm.rvs(mean,std, size=n)
pct_5 = stats.norm.ppf(.95,mean,std)
plt.hist(results,bins=10)
plt.axvline(pct_5)
plt.annotate(pct_5,xy=(pct_5,6))
plt.show()
```
